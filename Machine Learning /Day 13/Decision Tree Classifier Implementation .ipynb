# Independent and dependent features

```python
X = dataset.data
y = dataset.target
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train , y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

## Post-Pruning:

- Post-pruning, also known as "backward pruning," involves growing a full decision tree first and then removing or collapsing nodes in a bottom-up manner.
- Once the tree is fully grown, pruning techniques such as cost complexity pruning (also known as minimal cost-complexity pruning or weakest link pruning) are applied. This involves evaluating the impact of removing a subtree (or a single node) on the overall performance of the tree.
- Pruning decisions are typically guided by a pruning criterion, such as cross-validation error, to determine the optimal level of complexity for the tree.
- Post-pruning can be computationally more expensive than pre-pruning since it involves growing a full tree before pruning.

```python
## Post Pruning with max depth = 2

treeclassifier = DecisionTreeClassifier(criterion='entropy',max_depth=2)
treeclassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
## Prediction 

y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# PrePruning / Hyperparaemter Tuning for Large Dataset


- Pre-pruning involves setting constraints on the growth of the decision tree during the construction phase itself.
- Decision tree algorithms may employ various stopping criteria to control the size of the tree while it's being built. These criteria include limits on tree depth, minimum number of samples required to split a node, maximum number of leaf nodes, etc.
- By imposing these constraints during the tree construction process, pre-pruning prevents the tree from becoming overly complex and captures the most important patterns in the data without memorizing noise or outliers.

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
clf = GridSearchCV(DecisionTreeClassifier(), param_grid=parameter , cv=5 , scoring='accuracy')
```

```python
clf.fit(X_train , y_train)
```

```python
clf.best_params_
```

```python
y_pred_test= clf.predict(X_test)
```

```python
print(accuracy_score(y_test , y_pred_test))
```

```python
print(classification_report(y_test , y_pred_test))
```

In summary, pre-pruning sets constraints during the tree construction process to prevent overfitting, while post-pruning involves growing a full tree first and then removing unnecessary nodes to improve generalization performance. Both techniques aim to strike a balance between model complexity and predictive accuracy. The choice between pre-pruning and post-pruning often depends on the specific characteristics of the dataset and the computational resources available.
