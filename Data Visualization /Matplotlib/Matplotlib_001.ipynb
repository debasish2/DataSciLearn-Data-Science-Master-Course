```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
```

```python
x1 = np.linspace(1,10,200)
x1
```

```python
y1 = np.cos(x1)
```

```python
y1
```

```python
plt.plot(x1,y1)
```

```python
plt.plot(x1,y1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cos Function Graph")
plt.show()
```

```python
y2 = np.sin(x1)
plt.plot(x1,y2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is Sign Graph")
plt.show()
```

```python
m = np.random.rand(50)
n = np.random.rand(50)
colours = np.random.rand(50)
size = 500*np.random.rand(50)
plt.scatter(m,n, c =colours , s=size,alpha=.4, marker = '3')
plt.xlabel("This is my x-axis")
plt.ylabel("This is my y-axis")
plt.title("This is scatter plot")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
```

```python
df.head()
```

```python
df
```

```python
plt.scatter(df['Country'],df['Spending_USD'])
```

```python
plt.bar(df['Country'],df['Spending_USD'])
plt.xlabel('Country')
plt.ylabel('Spending_USD')
plt.title("Health")
plt.show()
```

```python
plt.barh(df['Country'],df['Spending_USD'])
```

```python
x3 = np.linspace(0,10,200)
y3 = np.sin(x3)
plt.figure(figsize=(5,2))
plt.plot(x3,y3,'--r')
plt.grid()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin graph")
plt.show()
```

```python
#Histogram
data = [1,1,1,4,4,5,8,8,9,10,5,4]
plt.hist(data)
plt.show()
```

```python
a = [4,5,6,7,8,9]
b = [60,30,80,90,45,50]
colours = ['pink','red','purple','green','#f7f4de','#8c8ff6']
plt.scatter(a,b , c=colours)
```

```python
# 3D

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
